```python
import tensorflow as tf
from keras.preprocessing import image
import numpy as np
import keras
from keras import Model
import time
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import Conv3D

fileName = "Trump.jpg"



# Global MobileNets model
mobile = keras.applications.mobilenet.MobileNet(weights="imagenet")
model = Model(mobile.input, mobile.layers[61].output)
```

```python
def write_to_file_weights_layer_Nineteen(fileName, W):
    fTemp = open(fileName, "w")
    for i in range(1):
        for j in range(1):
            for k in range(512):
                for l in range(512):
                    fTemp.write(str(W[i][j][l][k]) + "\n")
    fTemp.close()
```

```python
import math as m
layer60 = model.layers[60]
W = np.array(layer60.get_weights())
print(W.shape)

batchNorm = model.layers[60].get_weights()

fMean = open("data/NineteenthLayer/Nineteenth_Layer_Mean.txt", "w")
fSD = open("data/NineteenthLayer/Nineteenth_Layer_StanDev.txt", "w")
fGamma = open("data/NineteenthLayer/Nineteenth_Layer_Gamma.txt", "w")
fBeta = open("data/NineteenthLayer/Nineteenth_Layer_Beta.txt", "w")


for i in range(len(batchNorm[0])):
    fGamma.write(str(float(batchNorm[0][i])) + "\n")

for i in range(len(batchNorm[1])):
    fBeta.write(str(float(batchNorm[1][i])) + "\n")

for i in range(len(batchNorm[2])):
    fMean.write(str(float(batchNorm[2][i])) + "\n")

for i in range(len(batchNorm[3])):
    # The Value for Epsilon is assumed to be 0.001 in this implementation
    fSD.write(str(m.sqrt(float(batchNorm[3][i]) + 0.001)) + "\n")


print(batchNorm[0])
print(batchNorm[1])
print(batchNorm[2])
print(batchNorm[3])

fMean.close()
fSD.close()
fGamma.close()
fBeta.close()
```

```python
def checkFile_NineteenthLayer():
    fExp = open('data/NineteenthLayer/Nineteenth_Layer_Output.txt', 'r')
    fAct = open('data/NineteenthLayer/output.txt', 'r')
    
    cExp = fExp.readlines()
    cAct = fAct.readlines()
    counter = 0
    for i in range(len(cExp)):
         if int(round(float(cAct[i].strip()))) != int(round(float(cExp[i].strip()))):
            counter += 1
            if(counter < 30):
                print(str(cAct[i].strip()) + "--->" + str(cExp[i].strip()))
            
        
    print("Number of mismatch - " + str(counter))
    
    fExp.close()
    fAct.close()
    
        
```

```python
checkFile_NineteenthLayer()
```
